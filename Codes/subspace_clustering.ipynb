```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn import metrics
from selfrepresentation import ElasticNetSubspaceClustering
```

```python
dat = pd.read_excel("athletewise data.xlsx").drop('Athlete', axis=1)
dat.head()
```

Standardizing data

```python
std_df = ((dat - dat.mean())/dat.std())
std_df.head()
```

```python
model = ElasticNetSubspaceClustering(n_clusters=3,algorithm='lasso_lars',gamma=50).fit(std_df)
```

```python
np.unique(model.labels_, return_counts=True)
```

```python
sil = []
cal = []
dav = []
for n in range(2,21):
    model = ElasticNetSubspaceClustering(n_clusters=n,algorithm='lasso_lars',gamma=50).fit(std_df)

    labels = model.labels_
    sil.append(metrics.silhouette_score(std_df, labels)) #Ranges from -1 to 1 with higher values being better
    cal.append(metrics.calinski_harabasz_score(std_df, labels)) #Higher value is better
    dav.append(metrics.davies_bouldin_score(std_df, labels)) #Values closer to 0 are better
```

```python
print(sil)
print(cal)
print(dav)
```
